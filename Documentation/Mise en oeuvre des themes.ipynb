# Comment sont mis en oeuvre les thèmes dans l'application  
### Données brutes  
Elles sont localisées dans le fichier /parametres.yaml  
### Données concernant le thème actif    
Elles sont localisées dans le fichier /src/ui/theme_global.py  
L'exécution de la méthode definir_theme_global retourne le thème actif (appelé style)   
### Choix du thème actif  
Défini comme attribut de la classe AppContext (fichier context.py). L'attribut s'appelle theme_en_vigueur  
Exemple: self.theme_en_vigueur = {"palette": "palette_moderne", "police": "police_moderne_1"}  
Autres attributs de AppContext self.palette et self.police obtenus par utilisation conjointe de self.theme_en_vigueur et de parametres.py
